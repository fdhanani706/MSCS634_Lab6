```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
transactions = [
    ['milk', 'bread', 'butter'],
    ['beer', 'bread'],
    ['milk', 'bread', 'beer', 'butter'],
    ['bread', 'butter'],
    ['milk', 'butter'],
    ['beer', 'chips'],
    ['milk', 'bread', 'chips'],
    ['bread', 'butter', 'chips']
]
```

```python
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_array, columns=te.columns_)

df.head()
```

```python
item_counts = df.sum().sort_values(ascending=False)

sns.barplot(x=item_counts.index, y=item_counts.values)
plt.title("Most Frequent Items")
plt.xticks(rotation=45)
plt.show()
```

```python
co_occurrence = df.T.dot(df)

sns.heatmap(co_occurrence, annot=True, cmap="coolwarm")
plt.title("Item Co-occurrence Heatmap")
plt.show()
```

```python
apriori_itemsets = apriori(df, min_support=0.3, use_colnames=True)
apriori_itemsets.sort_values(by="support", ascending=False)
```

```python
top_apriori = apriori_itemsets.sort_values(by="support", ascending=False).head(5)

sns.barplot(x=top_apriori['itemsets'].astype(str), y=top_apriori['support'])
plt.title("Top Apriori Itemsets")
plt.xticks(rotation=45)
plt.show()
```

```python
fp_itemsets = fpgrowth(df, min_support=0.3, use_colnames=True)
fp_itemsets.sort_values(by="support", ascending=False)
```

```python
top_fp = fp_itemsets.sort_values(by="support", ascending=False).head(5)

sns.barplot(x=top_fp['itemsets'].astype(str), y=top_fp['support'])
plt.title("Top FP-Growth Itemsets")
plt.xticks(rotation=45)
plt.show()
```

```python
rules_apriori = association_rules(apriori_itemsets, metric="confidence", min_threshold=0.6)
rules_apriori
```

```python
sns.scatterplot(data=rules_apriori, x="confidence", y="lift")
plt.title("Apriori Rules: Confidence vs Lift")
plt.show()
```

```python
rules_fp = association_rules(fp_itemsets, metric="confidence", min_threshold=0.6)
rules_fp
```

Customers who buy milk and bread are likely to also buy butter.
High lift values indicate strong relationships beyond random chance.
These insights can help in product placement and promotions.

Comparison of Apriori vs FP-Growth:

- FP-Growth was faster because it avoids generating candidate itemsets.
- Apriori is easier to understand but computationally expensive.
- Both produced similar frequent itemsets for this small dataset.
